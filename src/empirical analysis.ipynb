# Image Processing Algorithms Empirical Analysis

## Importing Required Libraries

```python
import time
import random
from typing import Callable

import matplotlib.pyplot as plt
from PIL import Image

from algorithms.adaptive_gamma_correction import adaptive_gamma_correction
from algorithms.gamma_correction import gamma_correction
from algorithms.gaussian_blur import gaussian_blur as gb
from algorithms.histogram_equalization import histogram_equalization
from algorithms.mean_blur import mean_blur
from algorithms.sepia import sepia
from algorithms.sobel import sobel
```

## Defining a Function to Test Performance

```python
def test_performance(image_sizes: list[int], algorithm: Callable, color_mode: str = 'RGB') -> list[float]:
	times = []
	for size in image_sizes:
		# Determining the number of channels based on the color mode
		num_channels = 1 if color_mode == 'L' else 3
		
		# Generating a list of random pixel values for each channel
		random_pixels = [random.randint(0, 255) for _ in range(num_channels)]
		
		# Creating an image filled with the random pixel values
		img = Image.new(color_mode, (size, size), tuple(random_pixels))
		
		start_time = time.time()
		# Applying the algorithm on the image
		_ = algorithm(img)
		end_time = time.time()
		
		times.append(end_time - start_time)
	
	return times
```

## Running the Performance Test
Running each algorithm in a different cell helps debugging issues.

```python
image_sizes = [300, 600, 900, 1200, 1500]
```

```python
agc_times = test_performance(image_sizes, adaptive_gamma_correction)
```

```python
gc_times = test_performance(image_sizes, gamma_correction)
```

```python
gb_times = test_performance(image_sizes, gb)
```

```python
he_times = test_performance(image_sizes, histogram_equalization, 'L')
```

```python
mb_times = test_performance(image_sizes, mean_blur)
```

```python
sepia_times = test_performance(image_sizes, sepia)
```

```python
sobel_times = test_performance(image_sizes, sobel, 'L')
```

## Plotting the Results

### Adaptive Gamma Correction Performance

```python
plt.figure(figsize=(10, 6))
plt.plot(image_sizes, agc_times, marker='o')
plt.title('Execution Time vs Image Size for Adaptive Gamma Correction')
plt.xlabel('Image Size (pixels)')
plt.ylabel('Execution Time (seconds)')
plt.grid(True)
plt.text(
	0.02, 0.95, 'Time complexity: O(n)', transform=plt.gca().transAxes, verticalalignment='top'
)

plt.show()
```
